```python
import pandas as pd
import numpy as np

musics = pd.read_csv('filtered_track_df.csv')
musics.info()
```

```python
musics.rename(columns={"name": "songs_name"}, inplace=True)
```

```python
musics = musics[['uri', 'artists_name', 'songs_name', 'genres', 'lyrics']]
musics
```

```python
musics['lyrics'] = musics['lyrics'].replace(r'\r\n',  '', regex=True)
```

```python
musics.isnull().sum()
```

```python
musics.dropna(inplace=True)
```

```python
musics.duplicated(subset='lyrics').sum()
```

```python
musics[musics["songs_name"].isin(["Nervous"])]
```

```python
#menghapus lirik yang sama
musics.drop_duplicates(subset='lyrics', inplace=True)
```

```python
musics.duplicated(subset='lyrics').sum()
```

```python
musics[musics["songs_name"].isin(["Nervous"])]
```

```python
musics['artist_n_song'] = musics['artists_name'] + ' - ' + musics['songs_name']
```

```python
musics
```

```python
import ast

def convert(obj):
  L = []
  for i in ast.literal_eval(obj):
    L.append(i)
  return L

musics['genres'] = musics['genres'].apply(convert)
```

```python
#menghapus spasi
musics['genres'] = musics['genres'].apply(lambda x:[i.replace(" ", "") for i in x])
musics['artists_name'] = musics['artists_name'].apply(lambda x:x.replace(" ", ""))
musics.head()
```

```python
#memisahkan per kata
musics['lyrics'] = musics['lyrics'].apply(lambda x:x.split())
musics['artists_name'] = musics['artists_name'].apply(lambda x:x.split())
musics.head()
```

```python
musics['tags'] = musics['lyrics'] + musics['artists_name'] + musics['genres']
musics.head(10)
```

```python
new_df = musics[['uri', 'artist_n_song', 'tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'][0]
```

```python
import nltk
from nltk.stem.porter import PorterStemmer

ps = PorterStemmer()
def stem(text):
  y = []

  for i in text.split():
    y.append(ps.stem(i))

  return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
new_df['tags'][0]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df.head(10)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(new_df['tags']).toarray()
vectors
```

```python
cv.get_feature_names_out()
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[0])), reverse=True, key=lambda x:x[1])[1:6]
```

```python
def recommend(music):
  music_index = new_df[new_df['artist_n_song'] == music].index[0]
  distances = similarity[music_index]
  musics_list = sorted(list(enumerate(distances)), reverse=True, key=lambda x:x[1])[1:6]

  for i in musics_list:
    print(i[1])
    print(new_df.iloc[i[0]].artist_n_song)
```

```python
recommend('Coldplay - True Love')
```
